```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours_studied=np.random.randint(0,14,size=500)
marks_obtained=hours_studied*6+np.random.randint(-10,10,size=500)
marks_obtained=np.clip(marks_obtained,0,100)
array=np.vstack([hours_studied,marks_obtained])
print(array)
```

```python
array=array.T
```

```python
print(array)
```

```python
data = pd.DataFrame({
    'hours': hours_studied,
    'marks': marks_obtained
})
```

```python
data.head()
```

```python
data.describe()
```

```python
plt.scatter(data['hours'],data['marks'])
plt.xlabel("hours")
plt.ylabel("marks")
plt.show()
```

```python
!pip install scikit-learn
```

```python
X=data['hours']
Y=data['marks']
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
X_train = X_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='blue', alpha=0.6, label='Actual Data')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.title('Linear Regression: Hours Studied vs Marks Obtained')
plt.legend()
plt.show()
```

```python
import pickle
with open('linear_regression_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
